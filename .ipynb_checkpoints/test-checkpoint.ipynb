## SEE IMAGES

```python
import os
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from scipy.io import loadmat
from PIL import Image

# Directories for images and YOLO detections
BASE_DIR = "/home/jovyan/Projeto_PIV/data/1.1/CTownAirport_1.1"
IMAGES_DIR = os.path.join(BASE_DIR, "images")
YOLO_DIR = os.path.join(BASE_DIR, "yolo")

def plot_bounding_boxes(image_path, yolo_data, ax):
    """Overlay YOLO bounding boxes on an image."""
    # Load the image
    img = Image.open(image_path)
    ax.imshow(img)

    # Add bounding boxes
    for bbox in yolo_data['xyxy']:
        x_blc, y_blc, x_trc, y_trc = bbox
        width = x_trc - x_blc
        height = y_trc - y_blc
        rect = patches.Rectangle((x_blc, y_blc), width, height, linewidth=2, edgecolor='red', facecolor='none')
        ax.add_patch(rect)

def display_images_with_bboxes():
    """Display 5 images with YOLO bounding boxes."""
    # Select 5 image files
    image_files = sorted(os.listdir(IMAGES_DIR))[:5]
    fig, axes = plt.subplots(1, 5, figsize=(20, 10))

    for i, img_file in enumerate(image_files):
        # Load YOLO detection for the current image
        frame_idx = int(img_file.split('_')[1].split('.')[0])  # Extract frame index
        yolo_file = os.path.join(YOLO_DIR, f'yolo_{frame_idx:04d}.mat')

        # Load YOLO data
        if os.path.exists(yolo_file):
            yolo_data = loadmat(yolo_file)

            # Plot image with bounding boxes
            img_path = os.path.join(IMAGES_DIR, img_file)
            plot_bounding_boxes(img_path, yolo_data, axes[i])
            axes[i].set_title(f"Image: {img_file}")
        else:
            axes[i].imshow(Image.open(os.path.join(IMAGES_DIR, img_file)))
            axes[i].set_title(f"No YOLO data: {img_file}")
        
        axes[i].axis("off")

    plt.tight_layout()
    plt.show()

# Run the function
display_images_with_bboxes()
```
